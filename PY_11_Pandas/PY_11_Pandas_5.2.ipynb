```python
def get_suburb_type(suburb):
    exclude_list = ['N', 'S', 'W', 'E']
    suburb_list = suburb.split(' ')
    suburb_type = suburb_list[-1]
    if suburb_type in exclude_list:
        suburb_type = suburb_list[-2]
    return suburb_type
```

```python
suburb_types = melb_df['Suburb'].apply(get_suburb_type)
display(suburb_types)
```

```python
print(suburb_types.nunique())
```

```python
display(suburb_types.value_counts())
```

```python
popular_stypes =suburb_types.value_counts().nlargest(119).index
print(popular_stypes)
```

```python
melb_df['Suburb'] = suburb_types.apply(lambda x: x if x in popular_stypes else 'other')
display(melb_df['Suburb'])
```
